#Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df, hue = None,palette = 'hls')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)

sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen ,hue = None, palette = 'Set2')
```

From above graphs we can see that most of the buyers are Females and even the purchasing power of females are greater than men.

###### Age

```python
sns.countplot(data=df, x='Age Group', palette='hls', hue='Gender')
```

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)

sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age,hue = None, palette = 'hls')
```

#### From the above graph we can see that most of the buyers are of the Age Group between 26-35 female.

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x = 'State' , y = 'Orders' , data = sales_state,hue = None, palette = 'hls')
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x = 'State' , y = 'Amount' , data = sales_state,hue = None, palette = 'hls')
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df, palette='hls')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_marr = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status' , y = 'Amount' , data = sales_marr,hue = 'Gender', palette = 'hls')
```

#### From above graphs we can see that most of the buyers are married(women) and they have high purchasing power.


##### Occupation

```python
ax = sns.countplot(x = 'Occupation', data = df, palette = 'hls')

sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occu = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Occupation' , y = 'Amount' , data = sales_occu, palette = 'hls')
```

#### From above graph we can see that most of the buyers are working in IT,Aviation and Healthcare sector.

##### Product

```python
ax = sns.countplot(x = 'Product_Category', data = df, palette = 'hls')

sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_prod= df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category' , y = 'Amount' , data = sales_prod, palette = 'hls')
```

### From above graphs we can see that most are from Food, Clothing and Electronics category.

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID' , y = 'Orders' , data = sales_state, palette = 'hls')
```
